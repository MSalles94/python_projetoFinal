```python
from scripts.leitura_dados import ler_dados
dados=ler_dados().exportar_df()
```

## Visão Culinária

```python
#1

def top_rating(cusines='Italian',ascending=False):
    filtro=dados['main_Cuisines']==cusines
    aux_df=dados[filtro]
    aux_df=aux_df[['Restaurant_Name','Aggregate_rating']].drop_duplicates()
    aux_df=aux_df.sort_values(by='Aggregate_rating',ascending=ascending)
    return aux_df
top_rating(cusines='Italian',ascending=False)
```
