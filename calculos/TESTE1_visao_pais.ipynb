```python
from scripts.leitura_dados import ler_dados
dados=ler_dados().exportar_df()
```

## visão país

```python
# 1 e 2 , cidades e restaurantes registrados
aux_df=dados[['Country_name','City','Restaurant_ID']].drop_duplicates().groupby(['Country_name']).agg({'City':'nunique','Restaurant_ID':'nunique'}).reset_index()
aux_df=aux_df.rename(columns={'City':'Cidades','Restaurant_ID':'Restaurantes'}).sort_values(by=['Cidades','Restaurantes'],ascending=False)
aux_df
```

```python
# 3 pais com mais restaurantes com nível de preço igual a 4


aux_df=dados.loc[dados['Price_range']=='gourmet',['Restaurant_ID','Country_name']].drop_duplicates()
aux_df.groupby(['Country_name']).count().sort_values(by='Restaurant_ID',ascending=False)
```

```python
#4 país com maior quantidade de culinaria distinta

dados[['Country_name','main_Cuisines']].drop_duplicates().groupby(['Country_name']).count().sort_values(by='main_Cuisines',ascending=False)
```

```python
#5 país com maior quantidade de avaliação feita

dados[['Country_name','Restaurant_ID','Votes']].drop_duplicates().groupby(['Country_name'])[['Votes']].sum().sort_values(by='Votes',ascending=False)
```

```python
#6 País com maior quantidade de restaurantes que entregam

aux_df=dados.loc[dados['Is_delivering_now']==1,['Country_name','Restaurant_ID']].drop_duplicates()

aux_df=aux_df.groupby(['Country_name']).count().sort_values(by='Restaurant_ID',ascending=False)
aux_df
```

```python
#7  país com mais restaurantes que aceitam reservas
aux_df=dados.loc[dados['Has_Table_booking']==1,['Country_name','Restaurant_ID']].drop_duplicates()
aux_df=aux_df.groupby(['Country_name']).count().sort_values(by='Restaurant_ID',ascending=False)

aux_df
```

```python
#8 País com mais avaliações na média

aux_df=dados[['Country_name','Restaurant_ID','Votes']].drop_duplicates()

aux_df=aux_df.groupby(['Country_name'])[['Votes']].mean().sort_values(by='Votes',ascending=False)
aux_df
```

```python
#9 e 10 pais que possui na média a maior/menor nota média registrada

aux_df=dados[['Country_name','Votes','Aggregate_rating','Restaurant_ID']].drop_duplicates()

aux_df['nota']=aux_df['Votes']*aux_df['Aggregate_rating']

aux_df=aux_df.groupby(['Country_name']).sum().drop(columns='Restaurant_ID')

aux_df['nota']=aux_df['nota']/aux_df['Votes']

aux_df[['nota']].sort_values(by='nota',ascending=False)
```

```python
#11. média de preço do prado para dois por país

aux_df=dados[['Country_name','Restaurant_ID','Average_Cost_for_two']].drop_duplicates()

aux_df=aux_df.groupby(['Country_name'])[['Average_Cost_for_two']].mean().sort_values(by='Average_Cost_for_two',ascending=False)

aux_df
```
