```python
from scripts.leitura_dados import ler_dados
dados=ler_dados().exportar_df()
```

## VISÃO CIDADES

```python
#1
aux_df=dados[['City','Restaurant_ID']].drop_duplicates()
aux_df=aux_df.groupby(['City']).count().sort_values(by='Restaurant_ID',ascending=False)
aux_df
```

```python
#2
aux_df=dados[dados['Aggregate_rating']>4]
aux_df=aux_df[['City','Restaurant_ID']].drop_duplicates()
aux_df=aux_df.groupby(['City']).count().sort_values(by='Restaurant_ID',ascending=False)
aux_df
```

```python
#3
aux_df=dados[dados['Aggregate_rating']<2.5]
aux_df=aux_df[['City','Restaurant_ID']].drop_duplicates()
aux_df=aux_df.groupby(['City']).count().sort_values(by='Restaurant_ID',ascending=False)
aux_df
```

```python
#4
aux_df=dados[['City','Average_Cost_for_two']].drop_duplicates()
aux_df=aux_df.groupby(['City']).mean().sort_values(by='Average_Cost_for_two',ascending=False).round(2)
aux_df
```

```python
#5
aux_df=dados[['City','main_Cuisines']].drop_duplicates()
aux_df=aux_df.groupby(['City']).count().sort_values(by='main_Cuisines',ascending=False)
aux_df
```

```python
#6
aux_df=dados[dados['Has_Table_booking']==1]
aux_df=aux_df[['City','Restaurant_ID']].drop_duplicates()
aux_df=aux_df.groupby(['City']).count().sort_values(by='Restaurant_ID',ascending=False)
aux_df
```

```python
#7
aux_df=dados[dados['Is_delivering_now']==1]
aux_df=aux_df[['City','Restaurant_ID']].drop_duplicates()
aux_df.groupby(['City']).count().sort_values(by='Restaurant_ID',ascending=False)
```
