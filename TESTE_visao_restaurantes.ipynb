```python
from scripts.leitura_dados import ler_dados
dados=ler_dados().exportar_df()
```

## Visão Restaurantes

```python
#1
aux_df=dados[['Restaurant_Name','Votes']].drop_duplicates()
aux_df.sort_values(by='Votes',ascending=False)
```

```python
#2

aux_df=dados[['Restaurant_Name','Aggregate_rating']].drop_duplicates()
aux_df.sort_values(by='Aggregate_rating',ascending=False)
```

```python
#3
aux_df=dados[['Restaurant_Name','Average_Cost_for_two']].drop_duplicates()
aux_df.sort_values(by='Average_Cost_for_two',ascending=False)
```

```python
#4

aux_df=dados[dados['main_Cuisines']=='Brazilian']
aux_df=aux_df[['Restaurant_Name','Aggregate_rating']].drop_duplicates()
aux_df.sort_values(by='Aggregate_rating',ascending=True)
```

```python
#5

aux_df=dados[((dados['Country_name']=='Brazil')&
              (dados['main_Cuisines']=='Brazilian'))]
aux_df=aux_df[['Restaurant_Name','Aggregate_rating']].drop_duplicates()
aux_df.sort_values(by='Aggregate_rating',ascending=False)
```

```python
#6
aux_df=dados[['Restaurant_Name','Has_Online_delivery','Votes']].drop_duplicates()

aux_df=aux_df.groupby(['Has_Online_delivery']).mean(numeric_only=True)
aux_df
```

```python
#7

aux_df=dados[['Restaurant_Name','Has_Table_booking','Average_Cost_for_two']].drop_duplicates()

aux_df.groupby(['Has_Table_booking']).mean(numeric_only=True)

```

```python
#8

filtro=((dados['main_Cuisines'].isin(['Japanese','BBQ']))&
        (dados['Country_name']=='United States of America'))

aux_df=dados[filtro]
aux_df=aux_df[['Restaurant_Name','Country_name','main_Cuisines','Average_Cost_for_two']].drop_duplicates()

aux_df=aux_df.groupby(['Country_name','main_Cuisines']).mean(numeric_only=True)
aux_df
```
